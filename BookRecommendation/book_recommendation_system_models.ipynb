## Collaborative based filtering
dataset : rated_books.csv 
we need users who have rated more than 100 times 
We need books which has got more than 50 ratings.

```python
rated_books.shape
```

```python
readers = rated_books.groupby(['User-ID']).agg({'Book-Rating':'count'}).rename(columns={'Book-Rating':'num_rating'})
```

```python
readers['num_rating'].min()
```

```python
readers['num_rating'].max()
```

```python
readers.shape
```

```python
for i in [10,20,50,100,200,500,700,1000,4000]:
    print(f"{i} : {readers[readers['num_rating'] > i].shape}")
```

```python
active_readers = readers[readers['num_rating'] > 100].index
```

```python
filtered_users = rated_books[rated_books['User-ID'].isin(active_readers)]
```

```python
books = rated_books.groupby(['Book-Title']).agg({'Book-Rating':'count'}).rename(columns={'Book-Rating':'num_rating'})
```

```python
books.shape
```

```python
books['num_rating'].min()
```

```python
books['num_rating'].max()
```

```python
for i in [10,20,50,60,100,200,500,700]:
    print(f"{i} : {books[books['num_rating'] > i].shape}")
```

```python
active_books = books[books['num_rating'] > 50].index
```

```python
filtered_users_books = filtered_users[filtered_users['Book-Title'].isin(active_books)]
```

```python
filtered_users_books.shape
```

```python
filtered_users_books.head()
```

```python
filtered_users_books.duplicated().sum()
```

```python
pt = filtered_users_books.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

Here each book is a vector with 426 dimention
I need to calcutae distance between each book
at last For each book I will find 5 similar books with closest distance

#### using cosine similarity

```python
# python3.10 -m pip install sciket-learn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

Novel 1: Year of Wonders : index(623) :(no boo available for recommendation)
Novel 2 : Harry Potter and the Sorcerer's Stone (Book 1) : index(212)

```python
# row index of any book you name
np.where(pt.index=="Harry Potter and the Sorcerer's Stone (Book 1)")[0][0]
```

```python
# using this index of books and similarity score of the books, 
# get all the similarity score of this book with every other book 
#similarity_scores[1]

# enumarate this similarity score of book
#list(enumerate(similarity_scores[1]))

# sort this by similarity score in desending order and select 5 scores from it leaving first
# Since 1st similarity score shows the sililarity of the book with itself.
similar_items = sorted(list(enumerate(similarity_scores[212])),key=lambda x:x[1],reverse=True)
for i in similar_items:
    print(i)
```

```python
top_similar_items = []
for i in range(len(similar_items)):
    if 0.3 < similar_items[i][1] < 1:
        top_similar_items.append(similar_items[i])

top_similar_items
```

```python
if top_similar_items==[]:
        print('no book available for recommendation')
else:
       for i in top_similar_items:
            print(pt.index[i[0]])
```

```python
def recommend_book(book_name):
    # fetch row index
    row_index = np.where(pt.index==book_name)[0][0]
    # use this row index of book to find 
    # the index of 5 other most similar book
    similar_items = sorted(list(enumerate(similarity_scores[row_index])),key=lambda x:x[1],reverse=True)
    top_similar_items = []
    for i in range(len(similar_items)):
        if 0.3 < similar_items[i][1] < 1:
            top_similar_items.append(similar_items[i])
    # use these index to get Book-Title
    if top_similar_items==[]:
        print('no book')
    else:
         for i in top_similar_items:
             print(pt.index[i[0]])
```

```python

recommend_book('1st to Die: A Novel')
```
